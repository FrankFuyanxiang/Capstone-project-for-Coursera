Introduction: It semms that CBA is considered a low level league compared with NBA or European basketball, that is true.But many former NBA player admit that basketball there is really competitive and the skill level is not as low as people thought to be.Thus I determined to do a research based on the CBA advanced data and NBA advanced data to compare the difference between leagues using my newly developed data science methods.

Data resource: RealGM stat(advanced data)
We select top 100 players from each league ranked by PER rating

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib notebook
```

```python
df_cba=pd.read_csv("C:/Users/Frank0123/Desktop/cbastat.csv",encoding='utf-8')
df_nba=pd.read_csv("C:/Users/Frank0123/Desktop/nbastat.csv",encoding='utf-8')
```

```python
sns.regplot('ORtg','DRtg',data=df_nba,color='blue')
sns.regplot('ORtg','DRtg',data=df_cba,color='red')
plt.title('ORtg&DRtg CBA(red) vs NBA(blue)')
```

We can see from the above linear regression lines that when the offensive rating is at the same level, the defensive rating tends to be higher in CBA which indicates that being an excellent player in CBA, your defense is more important than NBAs.We just put NBA as a standardlized sample.It doesn't mean elite NBA players are poorer in defense, for they are compete in totally different environment,and NBA players are stronger and faster.Data shows that defense is really important in CBA compared to the money ball theory teams such as GSW or Rockets.In conclusion, if you want to be an elite player in CBA, play defense,therefore,that's why CBA is more "competitive".

```python
sns.regplot('AST%','TOV%',data=df_nba,color='blue')
sns.regplot('AST%','TOV%',data=df_cba,color='red')
plt.title('AST%&TOV% CBA(red) vs NBA(blue)')
```

As we all know, the AST/TOV number is a basic stat to estimate a player's passing skill,which is also known as organization ability.The data shows that the difference between CBA and NBA in passing is slight.We can combine our data analysis with eye test so that we can conclude that the offense patern and passing option in CBA is similar to NBA,especially solo and pick&roll.The foreign players in CBA act as all star in NBA,who carry the whole team in the last minutes of the game.

```python
sns.regplot('TRB%','BLK%',data=df_nba,color='blue')
sns.regplot('TRB%','BLK%',data=df_cba,color='red')
plt.title('TRB%&BLK% CBA(red) vs NBA(blue)')
```

That's really interesting.If we consider TRB% of an indication of paint zone dominance,then we will find that NBA player has much more power inside.The BLK% is greatly influenced by players height and jump height,which are obvious downsides for most Chinese players.There are more blocks in NBA,and when we connect it with pic1,we would realize that defensive player is really important and precious in today's CBA league environment.

```python
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
ax1.boxplot('USG%',data=df_cba)
ax2.boxplot('USG%',data=df_nba)
ax3.boxplot('PER',data=df_cba)
ax4.boxplot('PER',data=df_nba)
plt.subplots_adjust(wspace=0,hspace=0)
```

It shows that the difference between CBA elite players are larger than the gap between NBA elite players,which could be explained that the top tier CBA players come from NBA,and they are super stars among Chinese players who are mostly in the bottom of the box.Watch the median in subplot 1&2 , they are similar,while in subplot 3&4,NBA performs better than CBA.That menas under the same usage,NBA players could produce more, even they are under top class competion.
